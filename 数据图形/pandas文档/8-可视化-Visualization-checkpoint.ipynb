## plot() 基本的画图方法

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

对于DataFrame对象，plot()方法能方便地对所有列画图。

```python
df = df.cumsum()
```

```python
plt.figure() #重新画图
df.plot()
```

### 盒图 box plot

### 区域图 area plot

创建区域图的方法有 Series.plot.area(), DataFrame.plot.area()。区域图默认是堆积状态的，为了能够做出堆积的区域图，这就要求每列的值要么全部是正数要么全部是负数。

如果输入含有NaN，会被默认填充为0.如果你想画图前去掉这些NaN或者填充其他值，可以在作图前使用dataframe.dropna(), dataframe.fillna()方法。

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))
```

```python
df.plot.area()
```

如果想要画出不是堆积形态的区域图，使用参数stacked=False, Alpha值默认为0.5。

```python
df.plot.area(stacked=False)
```

### 散点图 Scatter Plot

可以使用DataFrame.plot.scatter()画散点图。散点图要求x、y轴的列都是数值类型。x、轴的设定可以直接通过参数x、y设定。

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=list('abcd'))
```

```python
df.plot.scatter(x='a', y='b')
```

一幅散点图中也可以画出多组散点图，方法很简单，使用ax参数即可。为了不混淆不同组的点，提醒一下使用color和label参数区别开。

```python
ax = df.plot.scatter(x='a', y='b', color='DarkBlue', label='Group 1')
df.plot.scatter(x='c', y='d', color='DarkGreen', label='Group 2', ax=ax)
```

颜色的设定 还可以通过scatter()方法中参数c 选择DataFrame中一列来设定颜色。

```python
df.plot.scatter(x='a', y='b', c='c', s=50)
```

```python
df.head()
```

scatter()还有很多参数，比如每个点的大小，可以通过某一列的值来设定每个点/气泡的大小。

```python
df.plot.scatter(x='a', y='b', s=df['c']*200)
```

### 六边形图 Hexagonal Bin Plot

创建六边形图的方法是DataFrame.plot.hexbin()。 六边形可以作为散点图的替代，特别是当数据大小分布过于密集，此时散点图很难区分各个点，六边形就是很好的一种方式。

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=list('ab'))
```

```python
df['b'] = df['b'] + np.arange(1000)
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
```

hexbin()方法中参数gridsize很有用，它控制x轴上六边形的个数，默认是100.gridsize越大，区间个数越多同时大小也越小。

### 饼图 Pie Plot

创建饼图的方法有 DataFrame.plot.pie()和Series.plot.pie()。如果数据中有NaN，作图前会被默认填充为0. 如果数据中有负值则会引起ValueError。

```python
series = pd.Series(3*np.random.rand(4), index=['a', 'b', 'c', 'd'], name='series')
```

```python
series.plot.pie(figsize=(6,6))
```

```python
series
```

DataFrame.plot.pie()要求通过y参数确定要作图的列或者subplots=True则会把每一列都作饼图。 

```python
df = pd.DataFrame(3 * np.random.rand(4,2), index=list('abcd'), columns=list('xy'))
```

```python
df.plot.pie(subplots=True, figsize=(8,4))
```

使用legend=False， 可以去掉说明。

使用labels和colors参数来定制每个扇区。

```python
series.plot.pie(labels=['AA', 'BB', 'CC', 'DD'], colors=['r', 'g', 'b', 'c'],
                    autopct='%.2f', fontsize=20, figsize=(6, 6))
```

如果pie()接收的数值总和小于1，会作出扇区。

```python
series = pd.Series([0.1] * 4, index=['a', 'b', 'c', 'd'], name='series2')
```

```python
series.plot.pie(figsize=(6, 6))
```

## 处理缺失值 

数据中难免会有缺失值，不同图形方法有不同的处理方法：


|    图形     | NaN处理方法  | 
| ------------- |:-------------:| 
| Line      | 在NaN处留有缺口| 
| Line(stacked)      | 填充0|   
| 条形图|填充0|    
|散点图|丢弃|
|直方图|丢弃|
|盒图|丢弃|
|区域图|填充0|
|KDE|丢弃|
|六边形|丢弃|
|饼图|填充0|



如果默认处理手段不是你想要的，你可以在作图前使用fillna()或者dropna()方法来自己处理。

### 并行坐标 Parallel Coordinates

并行坐标用于对多变量数据作图。可以观察聚类簇。 每条竖线代表一个属性。

```python
from pandas.tools.plotting import parallel_coordinates

df = pd.DataFrame(np.random.randn(100,3), columns=['A','B', 'C'])
df['X'] = np.random.choice(['Alpha', 'Beta', 'Theta'], size=100)
plt.figure()

parallel_coordinates(df, 'X')
```

### Lag图

**Lag图用于检查数据集或时间序列是否是随机的**。随机数据在Lag图中不会表现出任何的结构。而结构暗示着数据是非随机的。

```python
from pandas.tools.plotting import lag_plot
plt.figure()
```

```python
data = pd.Series(0.1 * np.random.rand(1000) +
        0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))
```

```python
lag_plot(data)
```

### 自相关图 Autocorrelation Plot 

自相关图通常用于检测时间序列中的随机性。通过计算数据在不同时间点的自相关。如果时间序列是随机的，自相关结果应该近似0。反之。

```python
from pandas.tools.plotting import autocorrelation_plot
```

```python
plt.figure()
```

```python
data = pd.Series(0.7 * np.random.rand(1000) +
       0.3 * np.sin(np.linspace(-9 * np.pi, 9 * np.pi, num=1000)))
```

```python
autocorrelation_plot(data)
```

### Bootstrap Plot

Bootstrap Plot用于可视化评估统计指标(比如平均值、中值等)的不稳定性。 通过从数据中随机选择固定大小的样本计算统计指标，然后不断重复此过程。

```python
from pandas.tools.plotting import bootstrap_plot
```

```python
data = pd.Series(np.random.rand(1000))
```

```python
bootstrap_plot(data, size=50, samples=500, color='grey')
```

### RadViz

RadViz是一种对多变量数据可视化的方法。基于一个简单的spring tension minimization algorithm.

```python
from pandas.tools.plotting import radviz
df = pd.DataFrame(np.random.randn(100,3), columns=['A','B', 'C'])
df['X'] = np.random.choice(['Alpha', 'Beta', 'Theta'], size=100)
plt.figure()
radviz(df, 'X')
```

### 误差线图 Error Bar

Series.plot()和DataFrame.plot()方法支持误差线图。 

```python
ix3 = pd.MultiIndex.from_arrays([['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b'], ['foo', 'foo', 'bar', 'bar', 'foo', 'foo', 'bar', 'bar']], names=['letter', 'word'])
```

```python
df3 = pd.DataFrame({'data1': [3, 2, 4, 3, 2, 4, 3, 2], 'data2': [6, 5, 7, 5, 4, 5, 6, 5]}, index=ix3)
gp3 = df3.groupby(level=('letter', 'word'))

means = gp3.mean()
```

```python
errors = gp3.std()
```

```python
means
```

```python
errors
```

```python
fig, ax = plt.subplots()
means.plot.bar(yerr=errors, ax=ax)
```
